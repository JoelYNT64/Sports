```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_goals_scored(team_df, team_name):
    """
    Plots the number of goals scored by a team, distinguishing between home and away games.

    team_df: DataFrame with columns ['Date', 'Goals_Scored', 'Home/Away']
    team_name: Name of the team for the title of the plot
    """
    team_df = team_df.copy()
    
    # Convert Date to datetime
    team_df['Date'] = pd.to_datetime(team_df['Date'])
    
    # Split home and away matches
    home_games = team_df[team_df['H/A'] == 'H']
    away_games = team_df[team_df['H/A'] == 'A']

    # Plot the data
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x= home_games['Date'], y= home_games['GF'], label='Home Goals')
    sns.scatterplot(x= away_games['Date'], y= away_games['GF'], label='Away Goals')

    # Labels and title
    plt.xlabel('Date')
    plt.ylabel('Goals Scored')
    plt.title(f'Goals Scored by {team_name} (Home vs Away)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    
    # Show the plot
    plt.show()
    
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x= home_games['Date'], y= home_games['GA'], label='Home Goals')
    sns.scatterplot(x= away_games['Date'], y= away_games['GA'], label='Away Goals')

    # Labels and title
    plt.xlabel('Date')
    plt.ylabel('Goals Allowed')
    plt.title(f'Goals Allowed by {team_name} (Home vs Away)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    
    # Show the plot
    plt.show()

# Example usage
# team_df = pd.read_csv("team_x.csv")  # Load the dataset
# plot_goals_scored(team_df, "Team X")
```

```python
plot_goals_scored(df, "Espagnol")
```
