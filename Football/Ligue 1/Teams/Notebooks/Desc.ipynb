```python
import pandas as pd
```

```python
df= pd.read_csv("../Data/Games.csv")
df
```

```python
df= df.drop(columns=["Comp"])
```

```python
df_teams = {tm: df[df["Team"] == tm] for tm in df["Team"].unique()}
```

```python
df_opps= {tm: df[df["Opp"] == tm] for tm in df["Opp"].unique()}
```

```python
for team in df_teams.keys():
    df_teams[team].to_csv(f"../Data/{team}.csv", index=False)
    df_opps[team].to_csv(f"../Data/vs {team}.csv", index=False)
```
