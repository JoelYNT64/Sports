```python
import pandas as pd
```

```python
df_sh= pd.read_csv("../Data/Shooting.csv")
#df_sh.drop(columns=['MP'], inplace= True)
df_sh
```

```python
df_rbd= pd.read_csv("../Data/Rebounding.csv")
df_rbd.drop(columns= ['MP'], inplace= True)
df_rbd
```

```python
df_bsc= pd.read_csv("../Data/Basic.csv")
df_bsc.drop(columns= ['MP'], inplace= True)
df_bsc
```

```python
df_adv_sh= pd.read_csv("../Data/Advanced Shooting.csv")
#df_bsc.drop(columns= ['MP'], inplace= True)
df_adv_sh
```

```python
df_adv_rbd= pd.read_csv("../Data/Advanced Rebounding.csv")
#df_bsc.drop(columns= ['MP'], inplace= True)
df_adv_rbd
```

```python
df_adv= pd.read_csv("../Data/Advanced.csv")
#df_bsc.drop(columns= ['MP'], inplace= True)
df_adv
```

```python
dfs= [df_sh, df_rbd, df_bsc, df_adv_sh, df_adv_rbd, df_adv]
```

```python
common_cols = set(dfs[0].columns)
df_sh.sort_values(by=["Team", "Opp", "Date"], inplace= True)
df_sh.reset_index(drop=True, inplace= True) 
for df in dfs[1:]:
    df.sort_values(by=["Team", "Opp", "Date"], inplace=True)
    df.reset_index(drop=True, inplace= True)
    common_cols &= set(df.columns)
common_cols = list(common_cols)  # Convert to list
common_cols
```

```python
df_cols= df_rbd[common_cols]
for df in dfs:
    df.drop(columns= common_cols, inplace= True)
dfs.append(df_cols)
```

```python
df_Teams_Games= pd.concat([df for df in dfs], axis=1)
df_Teams_Games
```

```python
cols = common_cols + [col for col in df_Teams_Games.columns if col not in common_cols]
df1= df_Teams_Games[cols]
df1
```

```python
id= ["Team", "Opp", "Date", "Result", "H/A", "Tm_PTS", "Opp_PTS"]
cols2= id+ [col for col in df1.columns if not(col in id)]
df2= df1[cols2].copy()
df2
```

```python
cols_to_delete= ["Rk"]
df2["Date"] = pd.to_datetime(df2["Date"], format="%Y-%m-%d")
df3 = df2.drop(columns= cols_to_delete)
df3
```

```python
df3["H/A"] = df3["H/A"].fillna("H").replace("@", "A")
df3 
```

```python
df3["Result"] = df3["Result"].str[0]
df3 
```

```python
df3.sort_values(by="Date", inplace=True)
df3
```

```python
df3["Tm_Rest"] = df3.groupby("Team")["Date"].diff().dt.days.fillna(5)
```

```python
df3
```

```python
df3["Opp_Rest"] = df3.groupby("Opp")["Date"].diff().dt.days.fillna(5)
df3
```

```python
df3['Tm_Pos']= 0.5* ((df3['Tm_FGA']+ 0.4* df3['Tm_FTA']- 1.07*(df3['Tm_ORB%']/100)* (df3['Tm_FGA']- df3['Tm_FG'])+ df3['Tm_TOV'])+ (df3['Opp_FGA']+ 0.4* df3['Opp_FTA']- 1.07*(df3['Opp_ORB%']/100)* (df3['Opp_FGA']- df3['Opp_FG'])+ df3['Opp_TOV']))
df3['Tm_Pos']
```

```python
df3['Tm_Pace']= 240/df3['MP']*df3['Tm_Pos']
df3['Tm_Pace']
```

Pace Calculation Formula=[240/(Team Minutes)]*(Possessionteam+Possessionopponent)/2

```python
df3['Opp_Pace']= 240/df3['MP']*df3['Tm_Pos']
df3['Opp_Pace']
```

```python
df_team = {comp_value: df3[df3["Team"] == comp_value] for comp_value in df3["Team"].unique()}
df_opp = {comp_value: df3[df3["Opp"] == comp_value] for comp_value in df3["Opp"].unique()}
```

```python
for team in df_team.keys():
    df_team[team].to_csv(f"../Data/{team}.csv", index=False)
    df_opp[team].to_csv(f"../Data/vs {team}.csv", index=False)
```
